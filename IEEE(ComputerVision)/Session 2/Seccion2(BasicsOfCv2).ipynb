# Packages

```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

# Dealing With Images


```python
Img1 = cv2.imread("The Data/Images/Wifi.jpg")
Img2 = cv2.imread("The Data/Images/fly.jpg")
```

## Show Images

```python
plt.subplot(1, 2, 1)
plt.imshow(Img1)

plt.subplot(1, 2, 2)
plt.imshow(Img2)

plt.show()
```

```python
cv2.imshow("1" ,Img1)
cv2.imshow("2" ,Img2)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

### Transformations

#### Resizing

```python
RImg1 = cv2.resize(Img1, (120, 120)) 
RImg2 = cv2.resize(Img2, (120, 120)) 
```

```python
cv2.imshow("1" ,RImg1)
cv2.imshow("2" ,RImg2)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

# Dealing With Videos

```python
Vid1 = cv2.VideoCapture("The Data/Videos/V1.mp4")
```

```python
while True:
    Succes, Frame = Vid1.read()
    
    if Succes == False:
        break
        
    RFrame = cv2.resize(Frame, (720, 520))
    cv2.imshow("1",RFrame)
    if cv2.waitKey(25) == ord(' '):
        break
cv2.destroyAllWindows()
```
