# Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

```python
ArrayG = np.ones((400,500) ) * 120
cv2.rectangle(ArrayG , (300 , 300) , (50,50) , (220) , -1)
```

```python
plt.imshow(ArrayG , cmap="gray" , vmin=0, vmax=255)
```

```python
ret, thresholded = cv2.threshold(ArrayG, 127, 255, cv2.THRESH_BINARY)
plt.imshow(thresholded , cmap = "gray" ,  vmin=0, vmax=255)
```

```python
sudoko = cv2.imread("../The_Data/Images/sudoku.png" ,cv2.IMREAD_GRAYSCALE)
coins = cv2.imread("../The_Data/Images/Coins.png" , cv2.IMREAD_GRAYSCALE)
```

 ### global thresholding

```python
ret, thresh_sudoko = cv2.threshold(sudoko, 127, 250, cv2.THRESH_BINARY)
```

```python
f , ( im1 , im2 ) = plt.subplots(1 ,2 ,sharey=True ,figsize = (10,5) )
f.subplots_adjust(wspace=.5)
im1.imshow(thresh_sudoko ,cmap = "gray" )
im1.set_title("thresh_sudoko")
im2.imshow(sudoko ,cmap = "gray" )
im2.set_title("sudoko")
```

Adaptive Threshold

```python
thresh_sudoko2 = cv2.adaptiveThreshold(sudoko, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 111, 2)
```

```python
f , ( im1 , im2 ) = plt.subplots(1 ,2 ,sharey=True ,figsize = (10,5) )
f.subplots_adjust(wspace=.5)
im1.imshow(thresh_sudoko2 ,cmap = "gray" )
im1.set_title("adaptiveThreshold")
im2.imshow(sudoko ,cmap = "gray" )
im2.set_title("sudoko")
```

Otsu Threshold

```python
ret, OTSU = cv2.threshold(coins, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
f , ( im1 , im2 ) = plt.subplots(1 ,2 ,sharey=True ,figsize = (10,5) )
f.subplots_adjust(wspace=.5)
im1.imshow(thresh_sudoko ,cmap = "gray" )
im1.set_title("OTSU")
im2.imshow(coins ,cmap = "gray" )
im2.set_title("coins")
```
